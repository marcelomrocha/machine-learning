**Importando as Bibliotecas**

```python
import sklearn
from matplotlib import pyplot
import seaborn as sns
import pandas as pd
import numpy as np

sns.set_theme()
```

##**Importando o Dataset e o transformando em um dataframe *Pandas***

```python
#criação do dataframe Pandas
df_iris = sns.load_dataset("iris")
print(df_iris.describe())
```

##**Obtendo a *média* de cada atributo usando a biblioteca *Pandas***

```python
media_sepal_length = df_iris["sepal_length"].mean()
print("Média de sepal_length: ", media_sepal_length)

media_sepal_width = df_iris["sepal_width"].mean()
print("Média de sepal_width: ", media_sepal_width)

media_petal_length = df_iris["petal_length"].mean()
print("Média de petal_length: ", media_petal_length)

media_petal_width = df_iris["petal_width"].mean()
print("Média de petal_width: ", media_petal_width)
```

##**Obtendo a *mediana* de cada atributo usando a biblioteca *Pandas***

```python
mediana_sepal_length = df_iris["sepal_length"].median()
print("Mediana de sepal_length: ", mediana_sepal_length)

mediana_sepal_width = df_iris["sepal_width"].median()
print("Mediana de sepal_width: ", mediana_sepal_width)

mediana_petal_length = df_iris["petal_length"].median()
print("Mediana de petal_length: ", mediana_petal_length)

mediana_petal_width = df_iris["petal_width"].median()
print("Mediana de petal_width: ", mediana_petal_width)
```

##**Calculando o *desvio* padrão de cada atributo usando a biblioteca *Pandas***

```python
desvio_sepal_length = df_iris["sepal_length"].std()
print("Desvio padrão de sepal_length: ", desvio_sepal_length)

desvio_sepal_width = df_iris["sepal_width"].std()
print("Desvio padrão de sepal_width: ", desvio_sepal_width)

desvio_petal_length = df_iris["petal_length"].std()
print("Desvio padrão de petal_length: ", desvio_petal_length)

desvio_petal_width = df_iris["petal_width"].std()
print("Desvio padrão de petal_width: ", desvio_petal_width)
```

##**Calculando *Q1* (primeiro quartil) de cada atributo usando a biblioteca *Pandas***

```python
q1_sepal_length = df_iris["sepal_length"].quantile(0.25)
print("Q1 de sepal_length: ", q1_sepal_length)

q1_sepal_width = df_iris["sepal_width"].quantile(0.25)
print("Q1 de sepal_width: ", q1_sepal_width)

q1_petal_length = df_iris["petal_length"].quantile(0.25)
print("Q1 de petal_length: ", q1_petal_length)

q1_petal_width = df_iris["petal_width"].quantile(0.25)
print("Q1 de petal_width: ", q1_petal_width)
```

##**Calculando Q3 (terceiro quartil) de cada atributo usando a biblioteca *Pandas***

```python
q3_sepal_length = df_iris["sepal_length"].quantile(0.75)
print("Q3 de sepal_length: ", q3_sepal_length)

q3_sepal_width = df_iris["sepal_width"].quantile(0.75)
print("Q3 de sepal_width: ", q3_sepal_width)

q3_petal_length = df_iris["petal_length"].quantile(0.75)
print("Q3 de petal_length ", q3_petal_length)

q3_petal_width = df_iris["petal_width"].quantile(0.75)
print("Q3 de petal_width: ", q3_petal_width)
```

##**Calculando a *obliquidade* de cada atributo usando a biblioteca *Pandas***

```python
obliquidade_sepal_length = df_iris["sepal_length"].skew()
print("Obliquidade de sepal_length: ", obliquidade_sepal_length)

obliquidade_sepal_width = df_iris["sepal_width"].skew()
print("Obliquidade de sepal_width: ", obliquidade_sepal_width)

obliquidade_petal_length = df_iris["petal_length"].skew()
print("Obliquidade de petal_length: ", obliquidade_petal_length)

obliquidade_petal_width = df_iris["petal_width"].skew()
print("Obliquidade de petal_width: ", obliquidade_petal_width)
```

##**Calculando a *curtose* de cada atributo usando a biblioteca *Pandas***

```python
curtose_sepal_length = df_iris["sepal_length"].kurtosis()
print("Curtose de sepal_length: ", curtose_sepal_length)

curtose_sepal_width = df_iris["sepal_width"].kurtosis()
print("Curtose de sepal_width: ", curtose_sepal_width)

curtose_petal_length = df_iris["petal_length"].kurtosis()
print("Curtose de petal_length: ", curtose_petal_length)

curtose_petal_width = df_iris["petal_width"].kurtosis()
print("Curtose de petal_width: ", curtose_petal_width)
```
